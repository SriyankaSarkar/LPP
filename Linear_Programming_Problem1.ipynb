##### ðŸ¤ºProblem:
\begin{align}
\textbf{Objective:} \quad \text{Maximize:} \quad Z & = 100x_1 + 40x_2 \\
\textbf{Subject to:} \quad 5x_1 + 2x_2 & \leq 1000 \\
3x_1 + 2x_2 & \leq 900 \\
x_1 + x_2 & \leq 500 \\
x_1, x_2 & \geq 0 \\
\end{align}


 ###### where,
\begin{aligned}
x_1 & : \text{ Represents the quantity of decision variable 1} \\
x_2 & : \text{ Represents the quantity of decision variable 2}
\end{aligned}

##### ðŸ“ŒGoal:
##### To find optimal values of x1 and x2 that :
\begin{array}{ll}
1. & \text{Maximize the value of } Z \\
2. & \text{Satisfy all given constraints:}
\end{array}
\begin{aligned}
5x_1 + 2x_2 &\leq 1000 \\
3x_1 + 2x_2 &\leq 900 \\
x_1 + x_2 &\leq 500 \\
x_1,\, x_2 &\geq 0
\end{aligned}



```python
# Import necessary libraries
import pandas as pd  # For data handling (if needed for extensions)
import numpy as np  # For numerical calculations
import matplotlib.pyplot as plt  # For graph plotting
import seaborn as sns  # For enhanced graph aesthetics
```

```python
# Install and import the 'pulp' library
# Make sure to install 'pulp' if not already done using pip
# pip install pulp
from pulp import *
```

```python
# Create an instance of an LPP model
prob = LpProblem("First_LP_Problem", LpMaximize)  # 'LpMaximize' indicates we are maximizing the objective function
```

```python
# Define decision variables with non-negativity constraints
x1 = LpVariable("x1", lowBound=0)  # x1 >= 0
x2 = LpVariable("x2", lowBound=0)  # x2 >= 0
```

```python
# Define the objective function to maximize profit
prob += 100 * x1 + 40 * x2, "Objective Function"
```

```python
# Add constraints
prob += 5 * x1 + 2 * x2 <= 1000.0, "1st Constraint"
prob += 3 * x1 + 2 * x2 <= 900.0, "2nd Constraint"
prob += x1 + x2 <= 500.0, "3rd Constraint"
```

```python
# Solve the problem
prob.solve()
```

```python
# Print results
print("Status:", LpStatus[prob.status])  # Status of the solution
```

```python
for v in prob.variables():
    print(v.name, "=", v.varValue)  # Optimal values for decision variables
```

```python
print("The optimal value of the objective function is =", value(prob.objective))
```

```python
# Plot the optimal solution
x=np.arange(0,700)
print(x)
plt.plot(x,500 - 2.5 * x,label='5x1 + 2x2 <= 1000')
plt.plot(x,450 - 1.5 * x,label='2x1+x2<=4')
plt.plot(x,250 - 0.5* x,label='6x1+2x2<=12')
plt.axis([0,600,0,700])
plt.legend()
plt.grid(True)
```
